```python
import pandas as pd
from sklearn import tree
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# define cols to load
part = 100000
desired_cols = ['DateOfCall', 'StopCodeDescription', 'PropertyType', 'FirstPumpArriving_AttendanceTime']

# load dataset, remove NaN
ds = pd.read_csv('data/lfb_incident.csv', usecols=desired_cols)
ds = ds.dropna()
ds = ds[0:part]
```

```python
ds[0:5]
```

```python
ds["DateOfCall"] = ds["DateOfCall"].apply(lambda x: datetime.strptime(x, '%d %b %Y').strftime('%Y-%m-%d'))
for col in ['DateOfCall', 'StopCodeDescription', 'PropertyType']:
    unique_date = ds[col].unique()
    for date in unique_date:
        ds[col] = ds[col].replace(date, np.where(unique_date == date)[0][0])
ds.to_csv('out.csv', index=False)
```
